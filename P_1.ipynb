```python
import numpy as np
```

```python
#Creating a Matrix of dimensions (300x80) where x = 300, y = 80
# B = np.zeros(shape=(300, 80))
# print(B)
# print(np.ndim(B))
x = 300
y = 80
B = np.empty((x, y))
B.fill(0)
print(B)
```

```python
#Creating a vector y of dimensions yx1 (i.e) (80, 1)
# v = np.zeros(shape = (80, 1))
# print(v)
v = np.empty((y, 1))
v.fill(0)
print(v)
```

```python
#Inserting Random Values in the Matrix
for i in range(x):
    for j in range(y):
        B[i][j] = np.random.randint(200)

print(B)
```

```python
#Inserting Normalised values in the vector
mu, sigma = 2, 0.01
for i in range(80):
    v[i] = np.random.normal(mu, sigma)

print(v)
```

```python
#Matrix Mulitplication
d = np.matmul(B, v)
print(d)

# d = []
# for i in range (len(B)):
#     row = []
#     for j in range(len(v[0])):
#         product = 0
#         for k in range(len(B[i])):
#             product += B[i][k] * v[k][j]
#         row.append(product)
#     d.append(row)
# d
            
```

```python
#Finding Mean and Standard Deviation
mean = np.mean(d)
print(mean)
stde = np.std(d)
print(stde)
```

```python
#Plotting
import matplotlib.pyplot as plt
plt.hist(d, bins=5)
plt.title("Random Numbers x Normal Distribution")
plt.show()
```
