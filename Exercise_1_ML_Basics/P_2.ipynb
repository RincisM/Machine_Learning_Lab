```python
import pandas as pd
```

```python
#Loading the Housing.csv file using pandas
data = pd.read_csv("./Housing.csv")
print(data.head)
```

```python
#Summarizing Each Field in the Data
print(data.describe())
```

```python
#Grouping Data by the field furnishingstatus and getting the average sqrt, price and bedrooms

grouped_data = data.groupby('furnishingstatus')[['area', 'price', 'bedrooms']].mean()
print(grouped_data)
```

```python
#Boxplotting each columnns (area, bedrooms, price)
import matplotlib.pyplot as plt
# boxplot = data.boxplot(column='area')
# boxplot.plot()
# plt.title('Area of the House')
# plt.show()

grouped_data.plot(kind = 'box', subplots=True, layout= (2,2), sharex=False, sharey=False)
plt.show()
```

```python
#Drawing a prediction line

import seaborn as sns

sns.lineplot(data = data, x = 'price', y ='area', hue='furnishingstatus')
plt.title('Price vs Area')
plt.show()
```
